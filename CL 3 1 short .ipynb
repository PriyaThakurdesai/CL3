```python
from xmlrpc.server import SimpleXMLRPCServer
import xmlrpc.client, threading

def fact(n):
    r = 1
    for i in range(1, n+1): r *= i
    return r

def run():
    s = SimpleXMLRPCServer(("localhost", 8000))
    s.register_function(fact, "factorial")
    print("RPC Server running on port 8000...")
    s.serve_forever()

threading.Thread(target=run, daemon=True).start()

s = xmlrpc.client.ServerProxy("http://localhost:8000/")
n = int(input("Enter number: "))
print("Factorial =", s.factorial(n))
```
