```python
import Pyro4, threading

@Pyro4.expose
class A:
    def add(self, x, y):
        return x + y

def run():
    d = Pyro4.Daemon()
    uri = d.register(A)
    print("URI:", uri)
    d.requestLoop()

# server background madhe
threading.Thread(target=run, daemon=True).start()
```
