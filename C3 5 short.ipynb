```python
import numpy as np, random
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

# Data
X=np.random.rand(100,3)
y=X[:,0]+X[:,1]+X[:,2]
Xtr,Xte,Ytr,Yte=train_test_split(X,y,test_size=0.2)

# Fitness
def f(n):
    m=MLPRegressor((n,),max_iter=100)
    m.fit(Xtr,Ytr)
    return m.score(Xte,Yte)

# GA
p=[5,10,15,8,12]
best=0; sol=0
history=[]

for g in range(20):
    s=[f(i) for i in p]
    m=max(s); x=p[s.index(m)]
    history.append(m)

    if m>best: best,sol=m,x

    print(f"Generation {g+1}: Best Neurons = {x}, Fitness = {m:.4f}")

    p=sorted(p,key=f,reverse=True)[:3]
    while len(p)<5:
        p.append(max(1,(p[0]+p[1])//2 + random.randint(-1,1)))

# Graph
plt.plot(range(1,21),history,marker='o')
plt.xlabel("Generation")
plt.ylabel("Best Fitness Value")
plt.title("GA–NN Optimization: Fitness vs Generation")
plt.show()

# Final
print("\nFinal Optimized Result")
print("Optimal Hidden Neurons:", sol)
print("Best Fitness Value:", round(best,4))
```
