```python
import random

# Servers (just names)
servers = ["Server-1", "Server-2", "Server-3"]
requests = 15


# 🔹 Round Robin
print("Round Robin:")
rr = {}
for i in range(requests):
    server = servers[i % len(servers)]
    rr[server] = rr.get(server, 0) + 1

print(rr)


# 🔹 Least Connections
print("\nLeast Connections:")
lc = {s: 0 for s in servers}

for i in range(requests):
    server = min(lc, key=lc.get)
    lc[server] += 1

print(lc)


# 🔹 Random Selection
print("\nRandom Selection:")
rand = {}

for i in range(requests):
    server = random.choice(servers)
    rand[server] = rand.get(server, 0) + 1

print(rand)
```
